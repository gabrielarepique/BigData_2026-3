#Calculadora de IMC

```python
print("*°*°* Calculadora de IMC *°*°*")
peso = float(input("Ingrese su peso en kilogramos (kg): "))
estatura = float(input("Ingrese su estatura en metros (m): "))

imc = peso / (estatura ** 2)

print(f"\nSu Índice de Masa Corporal (IMC) es: {imc:.2f}")

if imc < 18.5:
    clasificacion = "Bajo peso"
elif 18.5 <= imc <= 24.9:
    clasificacion = "Peso normal"
elif 25.0 <= imc <= 29.9:
    clasificacion = "Sobrepeso"
else:
    clasificacion = "Obesidad"

print(f"Clasificación: {clasificacion}")
```

#Sistema de notas estudiantes

```python
notas = []
for i in range(4):
    while True:
        try:
            nota = float(input(f"Ingrese la nota {i+1} (0.0 - 5.0): "))
            if 0.0 <= nota <= 5.0:
                notas.append(nota)
                break
            else:
                print("La nota debe estar entre 0.0 y 5.0. Intente de nuevo.")
        except ValueError:
            print("Entrada inválida. Por favor, ingrese un número.")

promedio = sum(notas) / len(notas)

print(f"\nEl promedio de las notas es: {promedio:.2f}")

if promedio >= 3.0:
    if 3.0 <= promedio <= 3.9:
        clasificacion_honor = "Aprobado"
    elif 4.0 <= promedio <= 4.4:
        clasificacion_honor = "Aprobado con mérito"
    else:
        clasificacion_honor = "Aprobado con honores"
    print(f"Estado: Aprobado ({clasificacion_honor})")
else:
    print("Estado: Reprobado")
```

##Ejercicio 4 (Tienda)

```python
inventario = {}

def agregar_producto():
    nombre = input("Nombre del producto: ").lower()
    if nombre in inventario:
        print("esa vaina ya esta")
        return
    cantidad = int(input("Cantidad inicial: "))
    precio = float(input("Precio unitario: $"))
    inventario[nombre] = [cantidad, precio]
    print(f"Producto '{nombre}' agregado")

def modificar_datos():
    nombre = input("Nombre del producto a modificar: ").lower()
    if nombre not in inventario:
        print("esa vaina no esta o no existe")
        return
    print(f"Datos actuales: Cantidad={inventario[nombre][0]}, Precio=${inventario[nombre][1]}")
    nuevo_precio = float(input("Nuevo precio: $"))
    inventario[nombre][1] = nuevo_precio
    print("Precio modificado")

def movimiento_mercancia():
    nombre = input("Nombre del producto: ").lower()
    if nombre not in inventario:
        print("no encontrado ")
        return

    op = input("¿Ingreso 'E' o Salida 'S'? ").upper()
    cantidad = int(input("Cantidad: "))

    if op == "E":
        inventario[nombre][0] += cantidad
        print(f"Ingreso registrado. Stock actual: {inventario[nombre][0]}")
    elif op == "S":
        if cantidad > inventario[nombre][0]:
            print("No hay suficiente stock")
        else:
            inventario[nombre][0] -= cantidad
            print(f"Salida registrada. Stock actual: {inventario[nombre][0]}")
    else:
        print("Opción inválida")

def eliminar_producto():
    nombre = input("Nombre del producto a eliminar: ").lower()
    if nombre in inventario:
        del inventario[nombre]
        print(f"Producto '{nombre}' eliminado ")
    else:
        print("Producto no encontrado ")

def mostrar_inventario():
    if not inventario:
        print("El inventario está vacío")
        return

    print("\n*°*°* INVENTARIO *°*°*")
    print(f"{'Producto':<15} {'Cantidad':<10} {'Precio':<10} {'Total':<10}")
    print("-" * 45)
    for nombre, datos in inventario.items():
        cantidad, precio = datos
        total = cantidad * precio
        print(f"{nombre:<15} {cantidad:<10} ${precio:<9.2f} ${total:<9.2f}")
    print("-" * 45)

while True:
    print("\n*°*°* SISTEMA DE INVENTARIO TIENDITA *°*°*")
    print("1. Agregar producto")
    print("2. Modificar datos")
    print("3. Ingreso/Salida de mercancía")
    print("4. Eliminar producto")
    print("5. Mostrar inventario")
    print("6. Salir")

    opcion = input("Elige una opción: ")

    if opcion == "1":
        agregar_producto()
    elif opcion == "2":
        modificar_datos()
    elif opcion == "3":
        movimiento_mercancia()
    elif opcion == "4":
        eliminar_producto()
    elif opcion == "5":
        mostrar_inventario()
    elif opcion == "6":
        print("Sistema cerrado cerradito")
        break
    else:
        print("ehh no revisa bien que opción estás escogiendo oye")
```

##Ejercicio 5 (Registro de estudiantes)

```python
estudiantes = {}

def agregar_estudiante():
    cedula = input("Ingrese la cédula del estudiante: ")
    if cedula in estudiantes:
        print("Error: Ya existe un estudiante con esa cédula.")
        return
    nombre = input("Ingrese el nombre del estudiante: ")
    edad = input("Ingrese la edad del estudiante: ")
    carrera = input("Ingrese la carrera del estudiante: ")
    estudiantes[cedula] = {
        "nombre": nombre,
        "edad": edad,
        "carrera": carrera
    }
    print(f"Estudiante {nombre} agregado correctamente.")

def modificar_estudiante():
    cedula = input("Ingrese la cédula del estudiante a modificar: ")
    if cedula not in estudiantes:
        print("Error: No se encontró un estudiante con esa cédula.")
        return

    print(f"Modificando estudiante: {estudiantes[cedula]['nombre']}")
    print("Deje en blanco un campo si no desea modificarlo.")

    nuevo_nombre = input(f"Nuevo nombre (actual: {estudiantes[cedula]['nombre']}): ")
    if nuevo_nombre:
        estudiantes[cedula]["nombre"] = nuevo_nombre

    nueva_edad = input(f"Nueva edad (actual: {estudiantes[cedula]['edad']}): ")
    if nueva_edad:
        estudiantes[cedula]["edad"] = nueva_edad

    nueva_carrera = input(f"Nueva carrera (actual: {estudiantes[cedula]['carrera']}): ")
    if nueva_carrera:
        estudiantes[cedula]["carrera"] = nueva_carrera

    print(f"Estudiante {cedula} modificado correctamente.")

def eliminar_estudiante():
    cedula = input("Ingrese la cédula del estudiante a eliminar: ")
    if cedula in estudiantes:
        del estudiantes[cedula]
        print(f"Estudiante con cédula {cedula} eliminado correctamente.")
    else:
        print("Error: No se encontró un estudiante con esa cédula.")

def buscar_estudiante():
    cedula = input("Ingrese la cédula del estudiante a buscar: ")
    if cedula in estudiantes:
        estudiante = estudiantes[cedula]
        print("\n--- Detalles del Estudiante ---")
        print(f"Cédula: {cedula}")
        print(f"Nombre: {estudiante['nombre']}")
        print(f"Edad: {estudiante['edad']}")
        print(f"Carrera: {estudiante['carrera']}")
        print("--------------------------------")
    else:
        print("Error: No se encontró un estudiante con esa cédula.")

def mostrar_todos_los_estudiantes():
    if not estudiantes:
        print("No hay estudiantes registrados.")
        return

    print("\n--- Lista de Estudiantes ---")
    for cedula, datos in estudiantes.items():
        print(f"Cédula: {cedula}, Nombre: {datos['nombre']}, Edad: {datos['edad']}, Carrera: {datos['carrera']}")
    print("----------------------------")


while True:
    print("\n--- Sistema de Gestión de Estudiantes ---")
    print("1. Agregar estudiante")
    print("2. Modificar estudiante")
    print("3. Eliminar estudiante")
    print("4. Buscar estudiante")
    print("5. Mostrar todos los estudiantes")
    print("6. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == '1':
        agregar_estudiante()
    elif opcion == '2':
        modificar_estudiante()
    elif opcion == '3':
        eliminar_estudiante()
    elif opcion == '4':
        buscar_estudiante()
    elif opcion == '5':
        mostrar_todos_los_estudiantes()
    elif opcion == '6':
        print("Saliendo del sistema.")
        break
    else:
        print("Opción inválida. Por favor, intente de nuevo.")
```

```python
import pandas as pd
import numpy as np
# -------- Parámetros --------
num_registros = 1000000  # puedes cambiarlo
np.random.seed(42)
# -------- Generación de dataset --------
productos = ["Café Expreso", "Café Americano", "Café Latte", "Capuchino", "Café Mocha"]
precios = [3000, 3500, 4000, 4500, 5000]  # en pesos colombianos
data = {
    "id_venta": np.arange(1, num_registros + 1),
    "producto": np.random.choice(productos, size=num_registros),
    "cantidad": np.random.randint(1, 6, size=num_registros),
    "precio_unitario": np.random.choice(precios, size=num_registros)
}
df = pd.DataFrame(data)
df.to_excel("ventas_cafe.xlsx", index=False) # Para exportar la BD en excel
```

```python
# -------- Agregar columna total por venta --------
df["total_venta"] = df["cantidad"] * df["precio_unitario"]
# -------- Visualización --------
print("Primeros registros del dataset:")
print(df.head())
# -------- Operaciones simples --------
print("\nCantidad total de unidades vendidas:", df["cantidad"].sum())
print("Total en compras (COP):", df["total_venta"].sum())
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum as spark_sum
import random
# -------- Inicializar Spark --------
spark = SparkSession.builder.appName("VentasCafe").getOrCreate()
# -------- Parámetros --------
num_registros = 100000000  # puedes cambiarlo
random.seed(42)
productos = ["Café Expreso", "Café Americano", "Café Latte", "Capuchino", "Café Mocha"]
precios = [3000, 3500, 4000, 4500, 5000]
# -------- Generación de dataset --------
data = [
    (i,
     random.choice(productos),
     random.randint(1, 5),
     random.choice(precios))
    for i in range(1, num_registros + 1)
]
columns = ["id_venta", "producto", "cantidad", "precio_unitario"]
df_spark = spark.createDataFrame(data, columns)
df_spark.show()

# Convertir en archivo .parquet
df_spark.write.parquet("ventas_cafe.parquet")
```

```python
# -------- Agregar columna total por venta --------
df_spark = df_spark.withColumn("total_venta", col("cantidad") * col("precio_unitario"))
# -------- Visualización --------
print("Primeros registros del dataset en Spark:")
df_spark.show(5)
# -------- Operaciones simples --------
resumen = df_spark.agg(
    spark_sum("cantidad").alias("total_unidades"),
    spark_sum("total_venta").alias("total_compras")
).collect()[0]
print("\nCantidad total de unidades vendidas:", resumen["total_unidades"])
print("Total en compras (COP):", resumen["total_compras"])
```
